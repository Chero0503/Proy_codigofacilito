Se desea presentar una visualización de datos del negocio con el objetivo de maximizar las ventas en el siguiente año. Se obtuvo una base de datos de la empresa la cual vamos a analizar.

# Observación de los Datos

Al observar los datos se muestra que tenemos 124658 filas y 28 columnas. Es una base de datos de ventas en la cual tenemos columnas importantes como la fecha de la venta, el producto con su respectivo id o identificador, la descripción del producto, el precio del producto.

Usando Phyton  con sus librerias pandas y numpy y analizando los datos podríamos sacar diferentes datos estadísticos pero primero hay que limpiar los datos.

Con la siguiente funcion ***dtypes*** vamos a revisar el nombre y el tipo de cada columna y eliminar algunas que no nos interesen.

```python
df.dtypes
```

#Limpieza de Datos

Al observar los datos, nos percatamos de que hay columnas que nos dan poca información o que no son de mucho interéscomo por ejemplo monto_abono, condicion_pago, domicilio y observaciones. Las cuales vamos a eliminar en el siguiente bloque de código utilizando la funcion **drop()**.

```python
nuevo_df= df.drop(columns = ['monto_abono','condicion_pago','colonia','domicilio','observaciones','observaciones']) 
nuevo_df
```

Vamos a eliminar los datos nulos para limpieza de los mismos.

```python
nuevo_df.dropna()
```

#Transformación de Datos

Quiero manipular el campo fecha_venta pero actualmente es de tipo Objet. Para poderlo usar lo cambiaremos de tipo usando la libreria **DateTime**.

```python
import datetime
from datetime import datetime #importamos la libreria datetime

nuevo_df['fecha_venta'] = pd.to_datetime(nuevo_df['fecha_venta'], errors='coerce')
nuevo_df.dtypes
```

Ahora fecha_venta ya es un campo tipo datetime.

```python
nuevo_df['Mes'] = nuevo_df['fecha_venta'].dt.month
nuevo_df['Año'] = nuevo_df['fecha_venta'].dt.year
nuevo_df.head(2) # El año y Mes aparecen en la última columna.
```

Como podemos observar, en el dataframe anterior se agregaron dos nuevas columas utilizando las funciones *dt.month* y *dt.year* de la libreria **DateTime**.

He querido manipular el campo estatus_venta, pero al parecer no se ha podido porque es de tipo Object, en el siguiente bloque se realizará el cambio del tipo de dato a **String**.

```python
nuevo_df['estatus_venta'] = nuevo_df['estatus_venta'].astype('string')
nuevo_df.dtypes
```

# Análisis de Datos

Una vez que los datos se limpiaron y transformaron podemos empezar con un pequeño análisis. Obteniendo la moda de las columnas se puede obtener cierta información.

```python
nuevo_df['id_producto'].mode()  
nuevo_df['descripcion_producto'].mode()
nuevo_df['estatus_cliente'].mode()
nuevo_df['lugar_trabajo'].mode()
nuevo_df['nombre_vendedor'].mode()
nuevo_df['Mes'].mode()
nuevo_df['Año'].mode()
```

Del código anterior podemos concluir lo siguiente:

*   El producto que más se vendió es el que tiene el id **14**.
*   El producto que más se ha vendido es el **Cobertor Jumbo Matrimonial**.
*   La mayoria de los clientes tienen un estatus **Regular**.
*   La ocupación mas frecuente del cliente es ser **Ama de casa**.
*   El vendedor que más vendio se llama **Felix**
*   El mes en el que mas se vende es **Junio**.
*   El año que más se ha vendido es en el **2022**.

De los datos anteriores podemos hacer un análisis del año 2022 que fue en el que mas se vendió. Por lo tanto quisiera saber cuanto dinero se obtuvo en el año 2022 con es estatus de las ventas igual a Pagado.

```python
nuevo_df.loc[(nuevo_df['estatus_venta'] == 'Pagada') & (nuevo_df['Año'] == 2022), 'monto_total'].sum()
```

HRMuebles recibió la cantidad de **$483,776.00**

Tambien quiero saber cuantos deudores estaban registrados en el mes de enero del 2021.

```python
nuevo_df.loc[(nuevo_df['estatus_venta'] == 'Deudora') & (nuevo_df['Mes'] == 1) & (nuevo_df['Año'] == 2021), 'monto_total'].sum()
```

HRMuebles tuvo **29850** deudores en enero del 2021

#Gráficos

Visualización de datos de las ventas que realizaron los trabajadores en el año 2022 y el producto que se vendió.

```python
año = nuevo_df[nuevo_df['Año']==2022]
```

```python
import altair as alt
import matplotlib.pyplot as plt

alt.Chart(año).mark_bar().encode(
    y=alt.Y("nombre_vendedor", title='Vendedor'),
    x=alt.X("descripcion_producto", title='Producto Vendido'),
    color=alt.Color("Mes", title="Mes del Año", scale=alt.Scale(scheme='reds')),
).properties(
    title='Ventas de productos realizadas por los vendedores en el 2022'
)

```

Se observa que solo hubo 7 vendedores activos en el 2022. 

```python

alt.Chart(año).mark_circle().encode(
    y=alt.Y("poblacion", title='Población'),
    x=alt.X("descripcion_producto", title='Producto Vendido'),
    color=alt.Color("Mes", title="Mes del Año", scale=alt.Scale(scheme='reds')),
).properties(
    title='Productos vendidos en las localidades del Sur de Puebla en el Año 2022'
)
```

Del gráfico anterior podemos concluir que en el 2022 se vendieron más productos en las localidades de Atlixco, Huaquechula, Cacaloxuchitl, Tepeojuma y Teacalco. Y que uno de los productos que más se vende en el suroeste del Estado de Puebla es la licuadora y el Cobertor Jumbo.
